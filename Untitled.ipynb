# Top 5 songs for the top artists

```python
def top_5_songs_for_an_artist(Artist_Name):
    top_5_songs = df[df['Artist Name'] == Artist_Name]['Song Name'].value_counts()[0:5]
    plt.figure(figsize=(10,5))
    sns.set(font_scale = 2)
    p = sns.barplot(x=top_5_songs.index, y= top_5_songs);
    p.set_ylabel("Song Frequency")
    p.set_title("Top 5 Favourite Songs of {}".format(Artist_Name), fontweight = 'bold' )
    plt.xticks(rotation=85);
    return
```

```python
Artist_Name = input('Enter Artist Name : ')
top_5_songs_for_an_artist(Artist_Name)
```
